Im folgenden Experiment werden wir uns mit der Klassifikation von Trüffelsorten beschäftigen. Dazu werden wir anhand von Daten  aus einer [Flüssigchromatographie mit Massenspektrometrie-Kopplung (LC-MS)](https://de.wikipedia.org/wiki/Fl%C3%BCssigchromatographie_mit_Massenspektrometrie-Kopplung) und mit Hilfe von Supervised Learning versuchen, die Trüffelsorten zu klassifizieren.

    Im Excelfile mit den Daten der LC-MS. Die ID (TM, TI, TA, TU) steht für die jeweilige Trüffelspezies.
- TM = [Tuber magnatum](https://de.wikipedia.org/wiki/Wei%C3%9Fe_Tr%C3%BCffel)
- TI = [Tuber indicum](https://en.wikipedia.org/wiki/Tuber_indicum)
- TA = [Tuber aestivum](https://de.wikipedia.org/wiki/Sommertr%C3%BCffel)
- TU = [Tuber uncinatum](https://de.wikipedia.org/wiki/Burgundertr%C3%BCffel)

Besonders die Tuber indicum und Tuber uncinatum sind sich sehr ähnlich und können nur schwer unterschieden werden. Qualität und Preis der Trüffel sind jedoch sehr unterschiedlich. Da die Tuber indicum deutlich günstiger sind, werden sie oft als Tuber uncinatum verkauft, da diese auch sehr gut den Geruch der Tuber uncinatum annehmen können. So können auch Experten getäuscht werden.
