# 3.7 Final Challenge

The Wold Anti-Doping Agency [WADA](https://www.wada-ama.org/en) made a raid in the team hotel of elite cyclist team.

They found syringes and vials with a clear liquid. They sent the vials to be analyzed in a lab. The lab found out, that liquid in the vials contains a virus and performed a sequence analysis. However, the lab is overrun with urgent tasks due to a new corona outbreak so they sent the results ([`sequencing_results.FASTQ`](https://github.com/jhumci/BLT_BDS/tree/main/data/doping)) to You for further analysis.

Your task is to create a report (written PDF-File) concerning the following topics. Upload the report and supplementing files in the sakai assignment (one participant per Group). The report must fit the formal requirements of the Academic Walkthrough (20% of the grading).

## 3.7.1 Sequence Quality Assurance (20%)

Given the following FASTQ-File with the raw reads ([`sequencing_results.FASTQ`](https://github.com/jhumci/BLT_BDS/tree/main/data/doping)). Evaluate the quality of the reads. How does the distribution of read length look like? What about the read quality in relation to the base position within the reads. What sampling technology was used to generate the data? What are the implications of the quality of the reads for the downstream analysis?

Minimal requiements for the submission:
- Chapter one of the report (`<lastname1>-<lastname2>-<lastname2>-report.pdf`)
- Two plots that provide an overview on the sequencing quality and a dicussion of the plots.

## 3.7.2 Sequence Assembly (20%)

Given the following FASTQ-File ([`sequencing_results.FASTQ`](https://github.com/jhumci/BLT_BDS/tree/main/data/doping)). Try to find out, what the cyclist where trying to do. First assabmle the sequence or as many contigs as possible from the reads (you do not have to write the algorithm yourself). Note, that You will have to store the results in a `FASTA` file if You want to use some external [tools](https://doua.prabi.fr/software/cap3) for assembly.

Next, use the [NCBI Database](https://blast.ncbi.nlm.nih.gov/Blast.cgi) to look for matches of Your sequences / contigs. (Hint: You are not only looking for a virus but also something from human specimens). What where the cyclists up to?

Minimal requiements for the submission:
- Chapter two of the report (`<lastname1>-<lastname2>-<lastname2>-report.pdf`)
- A FASTA-File of the assembled sequence(s) or contigs (`contigs.FASTA`)
- A screenshot of the first 10 matches You found to the sequences/ contig via blast (You do not have to search all for all the contigs, but You should find at gene of a virus and a human gene)
- A discussion on what the idea of the cyclist teamcould be using at least two scientific papers as sources


## 3.7.3 Gene Expression Analysis (20%)

The following [study](https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc=GSE186294) treated different subjects with a doping drug and compared their gene expression to the status before the drug administration. Your task is to find out, whether the treatment had any effects on the gene expression of the subjects.

You are provided with a cleaned overview on the samples ([`Illumia_clean_sample_data.csv`](https://github.com/jhumci/BLT_BDS/tree/main/data/doping)) and the raw reads of the gene expression data ([`GSE186294_Illumina_RNAseq_raw_counts_EnsemblID.txt`](https://github.com/jhumci/BLT_BDS/tree/main/data/doping)). You can base Your analysis on the observations `Base1` (before the drug was given) and `EPO4` (after the drug was given to the subjects).

Minimal requiements for the submission:
- Chapter three of the report (`<lastname1>-<lastname2>-<lastname2>-report.pdf`)
- Verbal decription of what You did in the code
- A table that describes what of the 48 samples belong to which subject and which point in time
- A table with log2-fold-change and p-values for the 10 most relevant genes (highest absolute fold change and significant p-value)
- Heatmap (scaled reads) for the 25 most relevant genes and all subjects 
- A plot that indicated how the expression of one of these genes developes over the course of the experiments (for the `Base1`, `Base2`, `EPO3`,`EPO4`, `Post7`) for each subject and in total
- Discussion of all tables and figures
